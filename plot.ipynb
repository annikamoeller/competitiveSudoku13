```python
import matplotlib.pyplot as plt
```

```python
# Initialize an empty dictionary
loaded_dict = {}

# Read lines from the text file
with open('experiment_results_with_score.txt', 'r') as file:
    for line in file:
        # Remove '{', '}', "'", and '\n' characters using replace
        clean_line = line.replace('{', '').replace('}', '').replace("'", '').replace('\n', '')
        key, value = clean_line.split(':')
        loaded_dict[key] = value
```

```python
sorted_dict_by_keys = dict(sorted(loaded_dict.items()))
sorted_dict_by_keys
```

```python
plt.plot(x_vals, easy2x2_greedy, label='greedy player', linestyle='-')
plt.plot(x_vals, easy2x2_random, label='random player', linestyle='--')
plt.plot(x_vals, easy2x2_ai2, label='A2 player', linestyle='solid')

plt.title("Easy 2x2 board")
plt.xlabel('Computation time allowed')
plt.ylabel('Win percentage')
plt.ylim(-10, 110)
plt.xticks([0.5, 1, 3, 5])
plt.legend()
```
